```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset (10 days)
days = list(range(1, 11))
opening = [100,103,105,104,107,110,108,106,109,112]
closing = [103,105,104,107,110,108,106,109,112,111]

# Calculate percentage change
pct_change = [(c - o) / o * 100 for o, c in zip(opening, closing)]

# Apply BUY / SELL / HOLD logic
signals = []
for i in range(len(pct_change)):
    today = pct_change[i]
    yesterday = pct_change[i-1] if i > 0 else None

    if today > 2 or (today > 0 and yesterday is not None and yesterday > 0):
        signals.append("BUY")
    elif today < -2 or (today < 0 and yesterday is not None and yesterday < 0):
        signals.append("SELL")
    else:
        signals.append("HOLD")

# Create DataFrame
data = pd.DataFrame({
    "Day": days,
    "Opening": opening,
    "Closing": closing,
    "Pct_Change": pct_change,
    "Signal": signals
})

print(data)
```

```python
#Line chart
plt.figure()
plt.plot(days, pct_change, marker='o')
plt.title("Line Plot of Daily Percentage Change")
plt.xlabel("Day")
plt.ylabel("% Change")
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.bar(days, pct_change)
plt.title("Daily Percentage Change (Bar Graph)")
plt.xlabel("Day")
plt.ylabel("% Change")
plt.show()
```

```python
plt.figure()
plt.hist(pct_change, bins=5)
plt.title("Histogram of Percentage Change")
plt.xlabel("% Change")
plt.show()
```

```python
plt.figure()
plt.scatter(opening, closing)
plt.title("Scatter Plot of Opening vs Closing")
plt.xlabel("Opening Price")
plt.ylabel("Closing Price")
plt.show()
```
